### Loops in lists

```python
my_list = ["apple", "banana", "cherry"]
my_list
```

```python
for fruit in my_list:
    print(fruit)
```

```python
# loop using index in list
for i in range(len(my_list)):
    print(my_list[i])
```

```python
# while loop
i=0
while i < len(my_list):
    print(my_list[i])
    i=i+1
```
