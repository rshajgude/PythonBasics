### Nameing convention for variable
1. Can not use reserved keywords.
2. You can not use any special character except _ in variable name.
3. Variable can have a-zA-Z0-9 and _ in it
4. Variable name can start with a-zA-Z and _
5. Rest all combiations are invalid variable names

```python
num1=1203
num1
```

```python
_name="Test Variable"
_name
```

```python
name^new="QA"
```

```python
first\name="Amit"
first
```
