```python
text="This is a sample\tText"
text.expandtabs()
```

```python
first_name.find("a")
```

```python
text="My name is {}"
text.format(first_name)
```

```python
text="My name is {}, and age is {}"
text.format(first_name, 34)
```

```python
text.format(34, first_name)
```

```python
# Named parameters formating
text="My name is {name}, and age is {age}"
age=34
text.format(name=first_name, age=age)
```

```python
text.format(age=age, name=first_name)
```

```python
first_name.index("a")
```

#### Diff between find and index is - 
Find will always return value there is no abort. If patrn not found it gives -1

Index abort execution when pattern is not found in string.
